### Code behind saving user data to local storage, retrieving and logging in Practice Log. This was the beginnings for the data portion of our project.
<script>
// Event listener for the save button click
document.getElementById("save-button").addEventListener("click", function () {
    // Get the practice time input value from the form
    const practiceTime = document.getElementById("practice-time").value;

    // Check if a valid practice time is entered
    if (practiceTime !== "") {
        // Get the current date in a readable format (e.g., "MM/DD/YYYY")
        const currentDate = new Date().toLocaleDateString();

        // Retrieve existing practice data from local storage, or initialize an empty object
        const practiceData = JSON.parse(localStorage.getItem("practiceData")) || {};

        // Update practice data with the current date and entered practice time
        practiceData[currentDate] = parseInt(practiceTime);

        // Save the updated practice data back to local storage
        localStorage.setItem("practiceData", JSON.stringify(practiceData));

        // Display an alert confirming the saved practice time
        alert(`Practice time (${practiceTime} minutes) saved for ${currentDate}`);

        // Clear the input field after saving
        document.getElementById("practice-time").value = "";

        // Refresh the practice log display
        displayWeeklyLog();
    } else {
        // Show an alert if no valid practice time is entered
        alert("Please enter a valid practice time.");
    }
});

// Function to display the weekly practice log
function displayWeeklyLog() {
    // Retrieve practice data from local storage, or initialize an empty object
    const practiceData = JSON.parse(localStorage.getItem("practiceData")) || {};

    // Get the table body element where the log will be displayed
    const tableBody = document.querySelector("#weekly-log table tbody");

    // Clear the existing content in the table body
    tableBody.innerHTML = "";

    // Iterate through each date in the practice data
    for (const date in practiceData) {
        // Create a new row in the table
        const row = tableBody.insertRow();

        // Create cells for date and practice time in the row
        const cellDate = row.insertCell(0);
        const cellTime = row.insertCell(1);

        // Populate the cells with date and practice time data
        cellDate.textContent = date;
        cellTime.textContent = practiceData[date];
    }
}

// Call the function to display the weekly practice log when the page loads
displayWeeklyLog();
</script>


### Moving Background
The moving background was also a big part of our project as it was incorporated on the tracker page itself.

- Originally we were going to go with this moving piano background (code shown below), but ended up replacing the moving background with a gif due to frame issues
  - Even if the frame was adjusted to fit the screen or the screen width of the particular device, each computer in our team showed a different display

<script>
// Get the canvas element and its 2D rendering context
const canvas = document.getElementById("alienWorld");
const ctx = canvas.getContext('2d');

// Constants for aspect ratio, maximum dimensions, and game speed
const ASPECT_RATIO = {{WIDTH}} / {{HEIGHT}};
const maxWidth = window.innerWidth;
const maxHeight = window.innerHeight;
var gameSpeed = 5;

// Calculate dimensions and position for the canvas
const canvasWidth = {{WIDTH}};
const canvasHeight = canvasWidth / ASPECT_RATIO;
const canvasLeft = 0; // Start from the left edge
const canvasTop = (maxHeight - canvasHeight) / 2;

// Set canvas dimensions and styles
canvas.width = canvasWidth;
canvas.height = canvasHeight;
canvas.style.width = `${canvasWidth}px`;
canvas.style.height = `${canvasHeight}px`;
canvas.style.position = 'absolute';
canvas.style.left = `${canvasLeft}px`;
canvas.style.top = `${canvasTop}px`;

// Define a Layer class to represent a scrolling background layer
class Layer {
  constructor(image, speedRatio) {
    this.x = 0;
    this.y = 0;
    this.width = {{WIDTH}};
    this.height = {{HEIGHT}};
    this.image = image;
    this.speedRatio = speedRatio;
    this.speed = gameSpeed * this.speedRatio;
    this.frame = 0;
  }
  // Update the layer's position based on speed
  update() {
    this.x = (this.x - this.speed) % this.width;
  }
  // Draw the layer on the canvas
  draw() {
    ctx.drawImage(this.image, this.x, this.y);
  }
}

// Create an instance of the Layer class for the background
const backgroundImg = new Image();
backgroundImg.src = '{{movingbg_csppassionproj}}';
var backgroundObj = new Layer(backgroundImg, 0.5);

// Function to update and draw the background layer
function background() {
  backgroundObj.update();
  backgroundObj.draw();
  requestAnimationFrame(background);
}

// Call the background function to start the animation loop
background();
</script>
